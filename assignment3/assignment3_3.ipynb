$$
Output = (Input - 1) \times stride - 2 \times padding
 + dilation \times (kernel size - 1) + output padding + 1
$$

$(8-1) \times 2 - 2 \times 1 + 1 \times (4-1) + 1 + 1 = 17$

The formula for calculating the BatchNorm is:

$$
y_i = \gamma \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}} + \beta
$$

Since we are using zero-mean and unit-variance, meaning that we need this normalized batch to have mean = 0 and std = 1

Formula will be:
$$
y_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$

assuming that $\epsilon$ is very small, we can ignore it, so the formula will be:
$$
y_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$

```python
import numpy as np
from torchaudio.transforms import SoudenMVDR


def normalization(batch, gamma=1, beta=0, eps=1e-5):
    mean = np.mean(batch)
    var = np.var(batch)
    normalized_batch = (gamma * (batch - mean) / np.sqrt(var + eps)) + beta
    return normalized_batch

mini_batch = np.array([6,8,10,6])
norm = normalization(mini_batch, eps = 0) # normally don't set eps = 0, but here we can set it to 0 to make sure the normalized batch has mean = 0 and std = 1
print("normalized batch:", norm)
```

The key mathematical difference between ReLU and LeakyReLU is that ReLU has a slope of 0 for negative inputs, while LeakyReLU has a small positive slope for negative inputs.

This means that LeakyReLU allows for a small amount of information to flow through the network even when the input is negative, which can help prevent the "dying ReLU" problem where neurons become inactive and stop learning.

Formula for ReLU:
$$
\text{ReLU}(x)=
\begin{cases}
x, & x>0 \\
0, & x\leq 0
\end{cases}
$$

Formula for LeakyReLU:
$$
\text{LeakyReLU}(x)=
\begin{cases}
x, & x>0 \\
\alpha x, & x\leq 0
\end{cases}
$$

where
$$
\alpha \in (0,1)
$$

LeakyReLU can help prevent the "dying ReLU" problem by allowing a small gradient to flow through the network even when the input is negative (LeakyReLU can aviods zero gradients for negative inputs). This can help keep neurons active and allow them to continue learning, which can improve the overall performance of the deep networks.
