```python
prob_A = 0.4
prob_B = 0.3

# (a) # due to A and B are independebt, P(A and B) = P(A) * P(B)
prob_A_and_B = prob_A * prob_B
print(f"(a) P(A and B) = {prob_A_and_B}")

# (b) P(A or B) = P(A) + P(B) - P(A and B)
prob_A_or_B = prob_A + prob_B - prob_A_and_B
print(f"(b) P(A or B) = {prob_A_or_B}")
```
