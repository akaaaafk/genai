### Question 1



```python
prob_A = 0.4
prob_B = 0.3

# (a) # due to A and B are independebt, P(A and B) = P(A) * P(B)
prob_A_and_B = prob_A * prob_B
print(f"(a) P(A and B) = {prob_A_and_B}")

# (b) P(A or B) = P(A) + P(B) - P(A and B)
prob_A_or_B = prob_A + prob_B - prob_A_and_B
print(f"(b) P(A or B) = {prob_A_or_B}")
```

### Question 2



```python
P_A = 0.5
P_B = 0.4

P_A_and_B = 0.5 * 0.4

P_A_given_B1 = P_A_and_B / P_B
print(f"P(A|B) = {P_A_given_B1}")
P_A_given_B2 = 0.7
print(P_A_given_B1 == P_A_given_B2)
```

Assume that A and B are independent, then $
P(A \mid B)=\frac{P(A \cap B)}{P(B)}
$ = (0.5 * 0.4) / 0.4 = 0.5, but the given P()A|B) = 0.7, so A and B are not independent.

### Question 3



```python
P_A = 0.6
P_B_given_A = 0.5
P_B_given_not_A = 0.2

P_A_and_B = P_B_given_A * P_A

P_B = P_B_given_A * P_A + P_B_given_not_A * (1 - P_A)

P_A_given_B = (P_A_and_B) / P_B
print(f"P(A|B) = {P_A_given_B}")
```

So P(A|B) $\approx$ 0.7895

### Question 4



```python
P_D = 0.02
P_not_D = 1 - P_D

P_positive_given_D = 0.95
P_negative_given_not_D = 0.90
P_positive_given_not_D = 1 - P_negative_given_not_D

P_positive = P_positive_given_D * P_D + P_positive_given_not_D * P_not_D

P_D_given_positive = P_positive_given_D * P_D / P_positive

print(f" Probability that a person who tests positive actually has the disease {P_D_given_positive}")
```

### Question 5





```python
# a
X = [85,90,95,100]
P = [0.375, 0.375, 0.125, 0.125]

E_X = sum(x * p for x, p in zip(X, P))
print(f" (a) Expected value E[X] = {E_X}")
```

```python
# b
E_X_squared = sum(x**2 * p for x, p in zip(X, P))
Var_X = E_X_squared - E_X**2
print(f" (b) Var(X) = {Var_X}")
```

```python
# c
X_2 = [85,90,85,95,90,85,100,90]
X_2_mean = np.mean(X_2)
print(f" (c) Mean of X_2 = {X_2_mean}")
```

Expected value from part (a) is 90, mean value of X_2 is also 90, so they are the same.

### Question 6



```python
# a
def information(x):
    return np.log2(x) if x != 0 else 0

def entropy_cal(probabilities):
    return -sum(p * information(p) for p in probabilities)

P = np.array([0.4,0.3,0.2,0.1])

entropy = entropy_cal(P)
print(f" (a) Entropy H(X) = {entropy}")
```

```python
# b
P_2 = np.array([0.25, 0.25, 0.25, 0.25])
entropy_2 = entropy_cal(P_2)
print(f" (b) Entropy H(X) for uniform distribution = {entropy_2}")
```
