# 1. uv package

为什么要用uv为包管理器

uv是一块现代化的python包与环境管理器，旨在为python开发者简化依赖环境，项目隔离以及构建流程

1. 简化的dependency management
- 统一的工作流：统一的工作流：与 pip 和 requirements.txt 不同，uv 采用声明式的 pyproject.toml 格式，将依赖跟踪、环境搭建和构建配置整合在单个文件中。
- 自动锁定locking：依赖项会自动锁定以确保确定性构建，使得项目在不同环境都可以具备复现性
- 更加便捷的增删操作：

uv add xxx
uv remove xxx

2. 环境隔离
- 集成的虚拟环境：uv会自动为项目创建并管理环境，就不用单独用venv或virtualenv了
- 无全局污染：所有项目依赖都隔离在.uv环境内，防止冲突

3. 提升开发效率
- 流线型初始化：用以下命令快速启动一个新项目

uv new project-name

4. 增强docker集成
- 简化docker构建：在用docker时，uv能够简化容器内的依赖安装和环境配置，与传统方法相比，用uv的典型dockerfile只需要更少的代码行数就可以搞定安装和配置
- 更少的依赖冲突：确定性的构建能确保您的 Docker 化应用即使在不同的机器上运行，也能保持行为一致。

5. future-proofing
- 兼容 PEP 517 和 PEP 518：uv 完全拥抱现代 Python 打包标准，使其与 Python 生态系统的最新进展保持兼容。
- 模块化与可扩展：该工具旨在随着现代开发者的需求不断演进，确保长期的高可用性。

# 2. fastapi



FastAPI是一个现代、快速（高性能）的Web框架，用于构建API，基于Python 3.6+，使用标准的Python类型提示。它的设计目标是让开发者能够快速地构建高性能的API，同时保持代码的简洁和可读性。

也就是一个用Python写Web API / 后端接口的框架，可以理解为：用Python快速写一个服务器，让别人或者前端通过网址来调用代码。比如有一个机器学习模型，想让别人通过网页/前端调用他，这就可以FastAPI包起来

使用 uv 包管理器安装 FastAPI
FastAPI 是一个现代、快速（高性能）的 Web 框架，用于构建 Python API。结合 uv 包管理器，设置 FastAPI 变得既简单又高效。

使用 uv 安装 FastAPI 的步骤
1. 初始化项目
首先，创建一个新的项目目录并使用 uv 进行初始化：

Bash
uv init sps_genai
这将会创建一个名为 sps_genai 的项目目录，其中包含一个由 uv 管理的预配置虚拟环境。

2. 进入项目目录
移动到新创建的项目目录中：

Bash
cd sps_genai
3. 安装 FastAPI
使用 uv 将 FastAPI 作为依赖项进行安装：

Bash
uv add fastapi --extra standard
4. 验证安装
确保 FastAPI 和 Uvicorn 已成功安装：

Bash
uv tree
该命令将显示已安装依赖项的列表，其中包含 FastAPI。

运行你的 FastAPI 应用程序
在 sps-genai 目录中创建一个名为 app/main.py 的文件，内容如下：

Python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello, FastAPI with UV!"}
使用 Uvicorn 运行该应用程序：

Bash
uv run fastapi dev
打开浏览器并访问 http://127.0.0.1:8000 以查看运行中的 API。

附加工具与功能
1. 自动生成文档
FastAPI 会自动生成交互式 API 文档。运行应用后，你可以访问以下地址：

Swagger UI: http://127.0.0.1:8000/docs

ReDoc: http://127.0.0.1:8000/redoc

2. 添加更多依赖项
如需添加其他依赖项（例如 Jupyter 或 Pydantic）：

Bash
uv add jupyter pydantic
3. 管理项目
uv 可以帮助你高效地管理虚拟环境和依赖项：

删除包： uv remove package-name

同步项目依赖与环境： uv sync

# 3. docker

docker是一个允许开发者在被称为容器（containers）的隔离环境中构建，共享和运行应用程序的平台。container容器可以轻松地将应用程序与其所有依赖项打包在一起，从而确保它们在不同的系统上都能可靠地运行。

docker好处：
- protability便捷性：在任何地方都能运行application，不用担心兼容性问题
- efficiency高效性：轻量，相比传统虚拟机
- scalability可拓展性：简化了应用程序的部署和拓展工作

docker = 单个容器管理工具
docker compose = 多个容器的一键编排工具

| 对比   | Docker        | Docker Compose                        |
| ---- | ------------- | ------------------------------------- |
| 管理对象 | 单个容器          | 多个容器                                  |
| 常用命令 | `docker run`  | `docker compose up`                   |
| 配置方式 | 命令行参数         | `compose.yaml` / `docker-compose.yml` |
| 适合场景 | 跑一个服务         | 跑一个完整项目                               |
| 例子   | 只跑 PostgreSQL | FastAPI + PostgreSQL + Redis          |

## Basic of Probability



P(A|B) = P(A and B) / P(B)

### Bayes Theorem

贝叶斯定理是概率论中的一个重要定理，它描述了在已知某些条件下，事件A发生的概率。它的公式如下：
P(A|B) = P(B|A) * P(A) / P(B)

### Entropy 熵

一个数据集合有多混乱，有多不确定

$$
Entropy(S) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

### Cross-Entropy 交叉熵
交叉熵是用来衡量两个概率分布之间的差异的一个指标，常用于分类问题中评估模型的性能。它的公式如下：
$$
CrossEntropy(p, q) = -\sum_{i} p(i) \log(q(i))
$$

### Kullback-Leibler Divergence KL散度
KL散度是用来衡量两个概率分布之间的差异的一个指标，常用于评估模型的性能。它的公式如下：
$$
KL(p || q) = \sum_{i} p(i) \log\left(\frac{p(i)}{q(i)}\right)
$$

## NLP modeling



NLP（自然语言处理）建模是指使用计算机算法和模型来理解、生成和处理人类语言的过程。NLP建模涉及多个步骤，包括文本预处理、特征提取、模型训练和评估等。常见的NLP模型包括：
1. 词袋模型（Bag of Words）：将文本表示为一个词汇表中每个词的出现频率。
2. TF-IDF（Term Frequency-Inverse Document Frequency）：一种改进的词袋模型，考虑了词频和逆文档频率。
3. Word Embeddings（词嵌入）：将词语映射到一个连续的向量空间中，常见的词嵌入方法有Word2Vec、GloVe等。
4. 循环神经网络（RNN）：适用于处理序列数据的神经网络，能够捕捉文本中的上下文信息。
5. 长短期记忆网络（LSTM）：一种特殊的RNN，能够更好地捕捉长距离依赖关系，适用于处理长文本。
6. Transformer：一种基于自注意力机制的模型，能够并行处理文本数据，适用于处理大规模文本数据，常用于预训练语言模型如BERT、GPT等。
